# CAO Points 2019 PDF

```python
# To use tabula, need to install first. Also requires java to be installed
import tabula
from tabula import read_pdf

url='http://www2.cao.ie/points/lvl8_19.pdf'

try: 
    df = read_pdf(url, pages='all')
    print(df)
except Exception as e:
    print('Error {}'.format(e))

    
# https://www.youtube.com/watch?v=6QSe_hlsUPc (30/10/2021)
```

```python
# Export to csv
tabula.convert_into(url, 'cao-2019.csv', output_format='csv', pages='all')

# https://pypi.org/project/tabula-py/
```
